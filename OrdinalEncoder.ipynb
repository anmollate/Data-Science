#Encoding the education and review column using ordinal encoding technique


```python
from sklearn.preprocessing import OrdinalEncoder

encoder=OrdinalEncoder(categories=[["HS","UG","PG"],["BAD","GOOD","EXCELLENT"]])
```

```python
x_encoded=encoder.fit_transform(df[["education","review"]])
```

```python
#conevrting back from np arrays to df
x_encoded=pd.DataFrame(x_encoded,columns=["education","review"])
```
