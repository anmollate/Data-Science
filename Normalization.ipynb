# Normalization

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("https://gist.githubusercontent.com/tijptjik/9408623/raw/b237fa5848349a14a14e5d4107dc7897c21951f5/wine.csv")
```

```python
df.head(5)
```

```python
x=df[["Alcohol","Malic.acid"]]
```

```python
y=df['Wine']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
#now normalizing the values
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train_scaled=scaler.transform(x_train)
x_test_scaled=scaler.transform(x_test)
#outputs in numpy array
```

```python
#converting the scaled values to df
x_train_scaled=pd.DataFrame(x_train_scaled,columns=['Alcohol','Malic.acid'])
x_test_scaled=pd.DataFrame(x_test_scaled,columns=['Alcohol','Malic.acid'])
```

```python
x_train_scaled
```

```python
x_train_scaled.describe()
```

```python
#training and testing the logistic regression model
from sklearn.linear_model import LogisticRegression
model1=LogisticRegression()
model2=LogisticRegression()

model1.fit(x_train, y_train)
model2.fit(x_train_scaled, y_train)

y_pred=model1.predict(x_test)
y_pred_scaled=model2.predict(x_test_scaled)

from sklearn.metrics import accuracy_score

print("Not scaled accuracy:",accuracy_score(y_test,y_pred))
print("Scaled accuracy:",accuracy_score(y_test,y_pred_scaled))
```
